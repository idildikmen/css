```python
import pandas as pd

# 1. Load your dataset
df = pd.read_csv('stakeholder_analysis_final.csv')

# 2. Define the list of columns to sum
```

```python
df.head()
```

```python
columns_to_sum = [
    "EU citizen", "Non-EU citizen", "Academic/research Institution",
    "Company/business", "Business association", "Non-governmental organisation (NGO)",
    "Other", "Trade union", "Environmental organisation", "Public authority",
    "Consumer organisation", "Prefer not to answer"
]

```

```python
# 3. Update only where the check is 0
mask = df['total_feedback_check'] == 0
df.loc[mask, 'total_feedback_check'] = df.loc[mask, columns_to_sum].sum(axis=1)

for col in columns_to_sum:
    # Create a new column name (e.g., "EU citizen (%)")
    percentage_col_name = f"{col} (%)"

    # Calculate the percentage
    # We use .fillna(0) in case the total is 0 to avoid "NaN" results
    df[percentage_col_name] = (df[col] / df['total_feedback_check'] * 100).fillna(0)
# 4. Save as CSV
```

```python
df.to_csv('updated_feedback_data_final.csv', index=False)
print("File saved successfully!")
```
